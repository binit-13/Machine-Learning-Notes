```python
import pandas as pd
import matplotlib.pyplot as plt
from string import ascii_uppercase
import random
import numpy as np

import timeit
```

```python
cols = list(ascii_uppercase)
```

```python
np.random.seed(12)
a = np.random.randint(1,100,size=(10000,10))
```

```python
df = pd.DataFrame(a,columns=cols[:10])
```

```python
def cross_tab(data):
  return pd.crosstab(data['A'],data['B'],data['C'],aggfunc=np.sum)

def group_by(data):
  return data.groupby('C')[['A','B']].sum()

def pivot(data):
  return pd.pivot_table(data=data,columns=data['B'], values=data['C'],index=data['A'],aggfunc='count',fill_value=0)
```

```python
func = [cross_tab, group_by]

measurement = []

for row in np.logspace(3,5,num=3):
  np.random.seed(12)
  a = np.random.randint(1,100,size=(int(row),10))
  data = pd.DataFrame(a,columns=cols[:10])

  for f in func:
    t = timeit.timeit(stmt='f(data)',globals=globals(),number=100)
    measurement.append({'function_name':f.__name__,'row_count': row,'duration':t})

print(measurement)
```

```python
m = pd.DataFrame(measurement)
print(m)
```

```python
m.groupby(['row_count', 'function_name'])['duration'].mean().unstack(level=0)
## level -1 le tyo inner most that is function name column ma aaucha where as 0 le outermost that is
## row_Count column ma gayo ani tyaha batta vayo
```
